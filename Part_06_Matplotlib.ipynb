**Brief history:** Matplotlib is a multiplatform data visualization library built on *NumPy* arrays, and designed to work with the broader *SciPy* stack. It was conceived by *John Hunter in 2002*, originally as a patch to IPython for enabling interactive MATLAB-style plotting via gnuplot from the IPython command line. [Read more](https://www.oreilly.com/library/view/python-data-science/9781491912126/ch04.html)

Hence this task happens to be the most important task in data sciences and Machine learning. *Matplotlib* has the ability to play well with many operating systems and graphics backends, and supports many output types. This happens to be one of its greatest strength.

**Importing Matplotlib:** This is done by importing the collection of functions that make matplotlib work, which is the **pyplot**. Each *pyplot* function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, and many more. To import a function in Matplotlib, we add a "." to call the function. For example, if you import the function pyplot (which is a collaction of functions), to import another function in pyplot you call it with another ".", that is: `import matplotlib.pyplot` to call the pyplot function in matplotlib, and `import matplotlib.pyplot.plot` to call the plot function.

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
```

We prepare some data: We creat some random data

```python
# create random data
np.random.seed(12) # We add this so that our random generation should not change each time we run the code
n = 30 # declare a variable n of value 30
x = [random.triangular() for i in range(n)] # Syntax: random.triangular(low, high, mode), returns returns a random floating number between the two specified numbers (both included)
y = [random.gauss(0.5, 0.25) for i in range(n)] # Syntax : random.gauss(mu, sigma), return a random floating point number with gaussian distribution.
colors = [random.randint(1, 4) for i in range(n)] # Syntax: randint(start, end), A random integer in range [start, end] including the end points.
areas = [math.pi * random.randint(5, 15)**2 for i in range(n)]
```

**Scatter plot:** A scatter plot is a type of plot that shows the data as a collection of points. The position of a point depends on its two-dimensional value, where each value is a position on either the horizontal or vertical dimension.

```python
# draw the plot
plt.figure()
plt.scatter(x, y)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple scatter plot")
plt.show()
```

**Buble plot**

```python
# draw the plot
plt.figure() # We use the default size figure
plt.scatter(x, y, label='bubles',s=areas, c=colors, alpha=0.85)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Buble plot")
plt.show()
```

**Connected scatter plot**

```python
#plt.plot(data, data, '-o')
# draw the plot
plt.figure()
plt.plot(x, y, '-o')
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("connected scatter plot")
plt.show()
```

**Line chart**

```python
# draw the plot
plt.figure() # We use the default size figure
plt.plot(x, y, label='bubles')#,s=areas, c=colors, alpha=0.85)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Line chart, not appropriate for this dataset")
plt.show()
```

We can observe that this chart is not appropriate for this dataset. Let us generate an appropriate dataset for this kind of plot.

```python
data = np.linspace(0, 10, 100)
#len(data), data
plt.figure() # We use the default size figure
plt.plot(data,data, label='data')
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Line chart")
plt.show()
```

**Subplots**: Below is an example of subplots

```python
fig, axs = plt.subplots(2, 2,figsize=(15,15))
axs[0, 0].scatter(x, y, label='bubles')
axs[0, 0].set_title('Axis [0, 0]: Scatter plot')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]: Line chart')
axs[1, 0].plot(x, y, '-o')
axs[1, 0].set_title('Axis [1, 0]: Connected scatter plot')
axs[1, 1].scatter(x, y, label='bubles',s=areas, c=colors, alpha=0.85)
axs[1, 1].set_title('Axis [1, 1]: customised scatter plot')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

**Histogram**

```python
# Remember that the generated y was such that it follows a guassian distribution, 
# we shall plot it's histogram and visualise it's probability density distribution.
import scipy.stats as st

fig, axs = plt.subplots(1, 2,figsize=(15,6))

axs[0].hist(y, density=True, bins=30, label="y")
axs[0].set_title('Histogram of y')
axs[0].legend(loc="upper left")
axs[0].text(0.8, 2.5, r'$\mu=0.5,\ \sigma=0.25$')
mn, mx = plt.xlim()
plt.xlim(mn, mx)
kde_xs = np.linspace(mn, mx, 301)
kde = st.gaussian_kde(y)
axs[1].hist(y, density=True, bins=30, label="y")
axs[1].plot(kde_xs, kde.pdf(kde_xs), label="PDF")
axs[1].legend(loc="upper left")
axs[1].set_title('Histogram of y with PDF line')
axs[1].text(0.7, 2.6, r'$\mu=0.5,\ \sigma=0.25$')
```

**Bar plots:** Because our data will not be appropriate for this plot, we shall creat a new sample of data to demonstrate this kind of plot.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['R', 'Scala', 'Spark', 'Python', 'Cython']
students = [40,37,35,49,50]
ax.bar(langs,students)
ax.set_title("Ideal when comparing several quantities and when changing one variable")
plt.show()
```

Suppose we have a data variable containing four series of five values. To illustrate this we shall construct a list of lists containing number of students enrolled in four fields of a Masters in Artificial intelligence training school over the past five years.

```python
data = [[33, 25, 60, 20,50],
[40, 26, 52, 17,32],
[35, 22, 45, 19,21],
[30, 23, 47, 59,65]]
X = np.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
xlables = ['0','2015', '2016', '2017', '2018', '2019']

ax.bar(X + 0.00, data[0], color = 'b', width = 0.35) #Each set of bar chart will be shifted 0.50 units from the previous one.
ax.bar(X + 0.50, data[1], color = 'g', width = 0.35)
ax.bar(X + 0.100, data[2], color = 'r', width = 0.35)
ax.bar(X + 0.150, data[3], color = 'y', width = 0.35)
ax.legend(labels=['ML', 'DL', 'NLP', 'AI Ethics'])
ax.set_xticklabels(xlables)
ax.set_ylabel('Number enrolled')
ax.set_ylabel('Years enrolled')
ax.set_title('Number of students enrolled in respective fields per year')
```

**Stacked barplots**: Given that we want to know the gender distribution of the various courses mentionned above and have the following new data separated according to gender. 

```python
N = 4

men = (110, 84, 40, 77)
women = (78, 83, 102, 147)
ind = np.arange(N)    # the x locations for the fields
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, men, width)
p2 = plt.bar(ind, women, width, bottom=men)

plt.ylabel('Enrolled students')
plt.ylabel('Fields')
plt.title('students by field and gender')
plt.xticks(ind, ('ML', 'DL', 'NLP', 'AI Ethics'))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

**Pie Chart**

```python
fig = plt.figure(figsize=(15,6))
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(students, labels = langs,autopct='%1.2f%%')
ax.set_title("Pie chart of our data on languages")
plt.show()
```

**Box plots:** Also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum. Best understood with the [illustration](https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51) below

We plot the boxplot of our data y generated previously

```python
from numpy.random import Generator, PCG64
rg = Generator(PCG64(1234))
y = rg.uniform(0, 50, 50)
np.random.seed(50) # We add this so that our random generation should not change each time we run the code
n=50
#y = [random.gauss(0.5, 0.25) for i in range(n)] 
fig = plt.figure()
# Create an axes instance
ax = fig.add_axes([0,0,1,1])
# Create the boxplot for our data y
bp = ax.boxplot(y)
#plt.annotate('Outlier', xy=(1.05, 0.07), xytext=(1.2, 0.07), arrowprops=dict(facecolor='green', shrink=0.01),)
plt.annotate('Min: least value excluding outliers', xy=(1.09, 0), xytext=(1.2, 0), arrowprops=dict(facecolor='yellow', shrink=0.01),)
plt.annotate('Q2', xy=(1.09, 42.5), xytext=(1.2, 42), arrowprops=dict(facecolor='black', shrink=0.01),)
plt.annotate('Median: 50% of data is greater than this value', xy=(1.09, 23.5), xytext=(1.2, 23), arrowprops=dict(facecolor='red', shrink=0.01),)
plt.annotate('Q1', xy=(1.09, 13), xytext=(1.2, 12.5), arrowprops=dict(facecolor='black', shrink=0.01),)
plt.annotate('Max: maximum value excluding outliers', xy=(1.09, 49.5), xytext=(1.2, 49), arrowprops=dict(facecolor='yellow', shrink=0.01),)
#plt.annotate('Outlier', xy=(1.05, 1.05), xytext=(1.17, 1.05), arrowprops=dict(facecolor='green', shrink=0.01),)
plt.title("Box plot of y")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
#plt.grid(True)
plt.show()
```
